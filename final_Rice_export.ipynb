LABELS MEANINGS :






HS CODE: Harmonised Commodity Description and Coding System and standardised format of classifying or describing type of shipments worldwide.

HS CODE DESCRIPTION :HS code is issued based on the classification of the shipment.

IMPORT VALUE FOB:Free on Board, or FOB is an Incoterm, which means the seller is responsible for loading the purchased cargo onto the ship, and all costs associated.

IMPORT VALUE CIF: CIF stands for Cost, Insurance, and Freight. These are the fees a seller pays to cover the costs, insurance, and freight of a dealer's order when it's enroute.

NET WEIGHT AND GROSS WEIGHT: Net weight refers to the weight of the goods only, while gross weight includes the weight of the goods, their packaging weight, and any additional accessories used for shipping.

TOTAL TEUS : A TEU (twenty-foot equivalent unit) is a measure of volume in units of twenty-foot long containers. For example, large container ships are able to transport more than 18,000 TEU (a few can even carry more than 21,000 TEU).

Perform a time-series analysis to identify trends or patterns in rice exports over the years.

```python
import pandas as pd

final_df['ARRIVAL DATE'] = pd.to_datetime(final_df['ARRIVAL DATE'])
final_df.set_index('ARRIVAL DATE', inplace=True)
```

```python
import matplotlib.pyplot as plt

# Plot the total quantity over time
plt.figure(figsize=(12, 6))
final_df['QUANTITY'].resample('M').sum().plot(title='Monthly Rice Exports Quantity')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.show()
```

Time-Series Decomposition:
Decompose the time series into trend, seasonality, and residual components.

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Decompose the time series
result = seasonal_decompose(final_df['QUANTITY'], model='additive', period=12)
result.plot()
plt.show()
```

```python
# Statistical Tests:
# Conduct statistical tests (e.g., Augmented Dickey-Fuller test) to check for stationarity.
```

```python
from statsmodels.tsa.stattools import adfuller

# Augmented Dickey-Fuller test
result_adf = adfuller(final_df['QUANTITY'])
print('ADF Statistic:', result_adf[0])
print('p-value:', result_adf[1])
print('Critical Values:', result_adf[4])
```

Autocorrelation and Partial Autocorrelation Analysis:
Analyze autocorrelation and partial autocorrelation to determine the order of the time series model (ARIMA).

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plot autocorrelation and partial autocorrelation
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 4))
plot_acf(final_df['QUANTITY'], lags=30, ax=ax1)
plot_pacf(final_df['QUANTITY'], lags=30, ax=ax2)
plt.show()
```

Model Building:
Build a time series model such as ARIMA (AutoRegressive Integrated Moving Average) using the identified parameters.

```python
from pmdarima import auto_arima
from statsmodels.tsa.arima.model import ARIMA

# Use auto_arima for automated parameter selection
auto_model = auto_arima(final_df['QUANTITY'], seasonal=False, suppress_warnings=True)
order = auto_model.order

# Fit ARIMA model using the selected parameters
model = ARIMA(final_df['QUANTITY'], order=order)
result_arima = model.fit()
print(result_arima.summary())
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Create a GeoDataFrame with world map shapefile
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Group by importer countries and count the occurrences
importer_counts = final_df['IMPORTER COUNTRY'].value_counts().reset_index()
importer_counts.columns = ['name', 'import_count']

# Group by exporter countries and count the occurrences
exporter_counts = final_df['COUNTRY OF ORIGIN'].value_counts().reset_index()
exporter_counts.columns = ['name', 'export_count']

# Merge the counts with the world GeoDataFrame using 'name' column
world = world.merge(importer_counts, how='left', on='name')
world = world.merge(exporter_counts, how='left', on='name')

# Plot the map
fig, ax = plt.subplots(1, 2, figsize=(15, 8))

# Plot importers
world.plot(column='import_count', cmap='OrRd', linewidth=0.8, ax=ax[0], edgecolor='0.8', legend=True)
ax[0].set_title('Importer Countries')

# Plot exporters
world.plot(column='export_count', cmap='BuGn', linewidth=0.8, ax=ax[1], edgecolor='0.8', legend=True)
ax[1].set_title('Exporter Countries')

plt.show()
```
